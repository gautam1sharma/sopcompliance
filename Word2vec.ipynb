```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-roberta-large-v1')
sentences = ["""The organization shall define an information security policy and supporting topic-specific policies, which shall be approved by management, made available to, and formally acknowledged by relevant personnel and interested parties. These policies shall be subject to review at scheduled intervals and upon the occurrence of any significant changes."""]
embeddings = model.encode(sentences)
print(embeddings.shape)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-roberta-large-v1')
sentences_1 = ["""Information security policy and topic-specific policies should be defined, approved by management,
published, communicated to and acknowledged by relevant personnel and relevant interested parties,
and reviewed at planned intervals and if significant changes occur."""]
embeddings_1 = model.encode(sentences_1)
print(embeddings_1.shape)
```

```python
# Similar sentences
#s1 = "The organization shall define an information security policy and supporting topic-specific policies, which shall be approved by management, made available to, and formally acknowledged by relevant personnel and interested parties. These policies shall be subject to review at scheduled intervals and upon the occurrence of any significant changes."
#s2 = " d, approved by management, published, communicated to and acknowledged by relevant personnel and relevant interested parties, and reviewed at planned intervals and if significant changes occur."

s1="The quick brown fox jumps over the lazy dog"
s2="A fast chestnut wolf leaps above a sluggish hound"
# Dissimilar sentence
s3 = "The sun orbits the Earth"

# Encode
e1 = model.encode([s1])
e2 = model.encode([s2])
e3 = model.encode([s3])

# Compare
def compare(e1, e2):
    cos_sim = np.dot(e1, e2.T) / (np.linalg.norm(e1) * np.linalg.norm(e2))
    l2 = np.linalg.norm(e1 - e2)
    return cos_sim.item(), l2.item()

print("Similar pair cosine sim / L2:", compare(e1, e2))
print("Dissimilar pair cosine sim / L2:", compare(e1, e3))
```
